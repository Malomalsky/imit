```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# Определим функцию интенсивности
def lambda_t(t):
    return 15 + 10 * np.sin(3 * np.pi * t / 1440)

# Определим максимальную интенсивность
lambda_max = 25

# Инициализируем время
t_i = 0
t_next = t_i

# Создадим список для хранения времен прибытия
arrivals = []

# Генерируем времена прибытия
while t_next < 24 * 60:  # продолжаем до конца суток
    U1 = np.random.uniform(0, 1)
    U2 = np.random.uniform(0, 1)
    t_next += -np.log(U1) / (60 / lambda_max)
    if U2 * lambda_max <= lambda_t(t_next):
        arrivals.append(t_next)


# Построение графика


sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
plt.plot(np.arange(0, 24 * 60, 1), [lambda_t(t) for t in np.arange(0, 24 * 60, 1)], label='Интенсивность λ(t)')
plt.scatter(arrivals, [lambda_t(t) for t in arrivals], color='red', s=10, alpha=0.5, label='Прибытия')
plt.xlabel('Время (минуты)')
plt.ylabel('Интенсивность')
plt.title('Нестационарный процесс Пуассона с интенсивностью λ(t) = 15 + 10sin(3πt/1440)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import quad
from scipy.optimize import root

# Определение функции интенсивности λ(t)
def lambda_t(t):
    return 15 + 10 * np.sin(3 * np.pi * t / 24)

# Определение интегральной функции L(x)
def L(x):
    return quad(lambda_t, 0, x)[0]

# Определение обратной функции x = L^(-1)(y)
def L_inv(y):
    return root(lambda x: L(x) - y, 0).x[0]

# Генерация случайного числа событий n
T = 24  # общее время моделирования
n = np.random.poisson(L(T))  # количество событий

# Определение времени наступления каждого события
arrival_times = [L_inv(np.random.uniform(0, L(T))) for _ in range(n)]

# Построение графика интенсивности λ(t)
t_values = np.arange(0, T, 0.1)
lambda_values = lambda_t(t_values)

sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
plt.plot(t_values, lambda_values, label='Интенсивность λ(t)')
plt.scatter(arrival_times, lambda_t(np.array(arrival_times)), color='red', s=10, alpha=0.5, label='Прибытия')
plt.xlabel('Время (часы)')
plt.ylabel('Интенсивность')
plt.title('Нестационарный процесс Пуассона с интенсивностью λ(t) = 15 + 10 * sin(3πt/24)')
plt.legend()
plt.show()
```
